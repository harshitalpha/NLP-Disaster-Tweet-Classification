Let's Explore some `MetaFeatures`

There can be many `Metafeatures` that we can explore. Like:
* `word_count`
* `char_count`
* `mean_word_len`
* `url_count`
* `punctuation_count`
* `hashtag_count`
* `mention_count`
* `unique_word_count`

````python
df_train['word_count'] = df_train['text'].apply(lambda x:len(str(x).split()))
df_test['word_count'] = df_test['text'].apply(lambda x:len(str(x).split()))

df_train['char_count'] = df_train['text'].apply(lambda x:len(str(x)))
df_test['char_count'] = df_test['text'].apply(lambda x:len(str(x)))

df_train['mean_word_len'] = df_train['text'].apply(lambda x: np.mean([len(w) for w in str(x).split()]))
df_test['mean_word_len'] = df_test['text'].apply(lambda x: np.mean([len(w) for w in str(x).split()]))


df_train['url_count'] = df_train['text'].apply(lambda x: len([w for w in str(x).split() 
                                                              if 'http' in w or 'https' in w]))
df_test['url_count'] = df_test['text'].apply(lambda x: len([w for w in str(x).split() 
                                                              if 'http' in w or 'https' in w]))


df_train['punctuation_count'] = df_train['text'].apply(lambda x: len([w for w in str(x) 
                                                                      if w in string.punctuation]))

df_test['punctuation_count'] = df_test['text'].apply(lambda x: len([w for w in str(x) 
                                                                      if w in string.punctuation]))

df_train['hashtag_count'] = df_train['text'].apply(lambda x : len([w for w in str(x).split()
                                                                  if '#' in w]))
df_test['hashtag_count'] = df_test['text'].apply(lambda x : len([w for w in str(x).split()
                                                                  if '#' in w]))

df_train['mention_count'] = df_train['text'].apply(lambda x : len([w for w in str(x).split()
                                                                  if '@' in w]))
df_test['mention_count'] = df_test['text'].apply(lambda x : len([w for w in str(x).split()
                                                                  if '@' in w]))

df_train['unique_word_count'] = df_train['text'].apply(lambda x: len(set(str(x).split())))
df_test['unique_word_count'] = df_test['text'].apply(lambda x: len(set(str(x).split())))
````

Conclusion from above plots:
* `Training` and `Test` set come from same distribution
* some metafeatures like `url_count`, `hashtag_count` and `mention_count` have no difference in distribution when saperated by target this means these features are not helpful in prediction terget
* other metafeatures have difference in distribution when saperated by target, this means these features can be helpful


````python
def generate_ngrams(text, n_gram=1):
    token = [token for token in text.lower().split(' ') if token != '' if token not in STOPWORDS]
    ngrams = zip(*[token[i:] for i in range(n_gram)])
    return [' '.join(ngram) for ngram in ngrams]

#UNIGRAMS

disaster_dict_uni = defaultdict(int)
non_disaster_dict_uni = defaultdict(int)

for text in df_train[df_train['target'] == 1]['text']:
    for gram in generate_ngrams(text,1):
        disaster_dict_uni[gram] += 1

for text in df_train[df_train['target'] == 0]['text']:
    for gram in generate_ngrams(text,1):
        non_disaster_dict_uni[gram] += 1

#BIGRAMS

disaster_dict_bi = defaultdict(int)
non_disaster_dict_bi = defaultdict(int)

for text in df_train[df_train['target'] == 1]['text']:
    for gram in generate_ngrams(text,2):
        disaster_dict_bi[gram] += 1

for text in df_train[df_train['target'] == 0]['text']:
    for gram in generate_ngrams(text,2):
        non_disaster_dict_bi[gram] += 1


#TRIGRAMS

disaster_dict_tri = defaultdict(int)
non_disaster_dict_tri = defaultdict(int)

for text in df_train[df_train['target'] == 1]['text']:
    for gram in generate_ngrams(text,3):
        disaster_dict_tri[gram] += 1

for text in df_train[df_train['target'] == 0]['text']:
    for gram in generate_ngrams(text,3):
        non_disaster_dict_tri[gram] += 1

````

````python
df_disaster_uni = pd.DataFrame(sorted(disaster_dict_uni.items(), key= lambda x:x[1])[::-1])
df_non_disaster_uni = pd.DataFrame(sorted(non_disaster_dict_uni.items(), key= lambda x:x[1])[::-1])

df_disaster_bi = pd.DataFrame(sorted(disaster_dict_bi.items(), key= lambda x:x[1])[::-1])
df_non_disaster_bi = pd.DataFrame(sorted(non_disaster_dict_bi.items(), key= lambda x:x[1])[::-1])

df_disaster_tri = pd.DataFrame(sorted(disaster_dict_tri.items(), key= lambda x:x[1])[::-1])
df_non_disaster_tri = pd.DataFrame(sorted(non_disaster_dict_tri.items(), key= lambda x:x[1])[::-1])
````

````python
fig,axes = plt.subplots(ncols = 2, figsize = (20,50))
sns.barplot(y = df_disaster_uni[0][:100], x = df_disaster_uni[1][:100], color="red", ax = axes[0])
sns.barplot(y = df_non_disaster_uni[0][:100], x = df_non_disaster_uni[1][:100], color="pink", ax = axes[1])
plt.show()
````

Lets Do text cleaning

### Removing URLs

````python
def remove_url(text):
    return re.sub(r"https?:\/\/t.co\/[A-Za-z0-9]+","",text)

df_train['text'] = df_train['text'].apply(lambda x:remove_url(x))
df_test['text'] = df_test['text'].apply(lambda x:remove_url(x))
````

### Removing HTML tags

````python
def remove_html(text):
    return re.sub(r"<.*?>","",text)
df_train['text'] = df_train['text'].apply(lambda x:remove_html(x))
df_test['text'] = df_test['text'].apply(lambda x:remove_html(x))
````

### Removing Punctuation

````python
def remove_punctuation(text):
    table=str.maketrans('','',string.punctuation)
    return text.translate(table)
df_train['text'] = df_train['text'].apply(lambda x:remove_punctuation(x))
df_test['text'] = df_test['text'].apply(lambda x:remove_punctuation(x))
````

# GLOVE FOR VECTORIZATION

### Create Word Corpus

````python
def create_corpus(df):
    vocab = {}
    for tweet in tqdm(df['text']):
        for word in tweet.split():
            try:
                vocab[word] += 1
            except KeyError:
                vocab[word] = 1
    return vocab

df = pd.concat([df_train[['id','text']],df_test[['id','text']]])
vocab = create_corpus(df)
````

### Load Embedding

````python
glove_embedding = {}

with open("Glove Embedding/glove.twitter.27B/glove.twitter.27B.100d.txt", 'r') as f:
    for line in tqdm(f):
        value = line.split()
        word = value[0]
        vector = np.asarray(value[1:],"float32")
        glove_embedding[word] = vector
````

### Getting Vocabulary And Text Coverage

````python
covered = {}
n_covered = 0
oov = {}
n_oov = 0

for word in tqdm(vocab.keys()):
    try:
        covered[word] = glove_embedding[word]
        n_covered = n_covered + vocab[word]
    except KeyError:
        oov[word] = vocab[word]
        n_oov = n_oov + vocab[word]

vocab_coverage = len(covered)/len(vocab)
text_coverage = n_covered/(n_covered + n_oov)
sorted_oov = sorted(oov.items(), key = lambda x:x[1])[::-1]

print("Text coverage : {}".format(text_coverage))
print("Vocab coverage : {}".format(vocab_coverage))
````

````python
embedding_matrix = np.zeros((len(word_indexes)+1,100))

for word in tqdm(word_indexes.keys()):
    if word_indexes[word] > len(word_indexes):
        continue
    emb_vec = glove_embedding.get(word)
    if emb_vec is not None:
        embedding_matrix[word_indexes[word]] = emb_vec
````

````python
embedding_matrix.shape
````

````python
def recall_m(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
    recall = true_positives / (possible_positives + K.epsilon())
    return recall

def precision_m(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
    precision = true_positives / (predicted_positives + K.epsilon())
    return precision

def f1_m(y_true, y_pred):
    precision = precision_m(y_true, y_pred)
    recall = recall_m(y_true, y_pred)
    return 2*((precision*recall)/(precision+recall+K.epsilon()))
````

````python
#model = Sequential()

inp = Input(shape = (31,))
embedding_layer = Embedding(len(word_indexes)+1,100,
                            embeddings_initializer=Constant(embedding_matrix),
                           input_length=padded_sequence.shape[1],
                           trainable = False)(inp)

x = SpatialDropout1D(0.3)(embedding_layer)
lstm1 = Bidirectional(LSTM(128,dropout = 0.2, return_sequences = True))(x)
lstm2 = Bidirectional(LSTM(128,dropout = 0.2, return_sequences = True))(lstm1)
lstm3 = Bidirectional(LSTM(128,dropout = 0.2, return_sequences = True))(lstm1)
add1 = Add()([lstm2,lstm3])
lstm4 = Bidirectional(LSTM(128,dropout = 0.3))(add1)
d1 = Dense(64, activation='relu')(lstm4)
d2 = Dense(64, activation='relu')(d1)
out = Dense(1,activation='sigmoid')(d2)

# model.add(embedding_layer)
# model.add(SpatialDropout1D(0.3, name = "x"))
# model.add(Bidirectional(LSTM(128,dropout = 0.2, return_sequences = True)))
# model.add(Bidirectional(LSTM(128,dropout = 0.2, return_sequences = True)))
# model.add(Add()([x,lstm2]))
# model.add(lstm3)
# model.add(d1)
# model.add(d2)
# model.add(add2)
# model.add(out)

model = Model(inputs=inp, outputs=out)

optimzer=Adam(learning_rate=1e-3)

model.compile(loss = "binary_crossentropy", optimizer=optimzer, metrics=['accuracy',f1_m,precision_m, recall_m])

model.summary()
````
